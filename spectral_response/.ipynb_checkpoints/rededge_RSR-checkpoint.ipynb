the QE table ends at wavelength 1099, but the filter sensitivity table ends at wavelength 900

```python
qe_l = qe.values[:,0]
filt_l = filt_trans.values[:,0]
```

```python
# qe_l, temp_arr[:,0]
qe_sub = qe.loc[(qe_l >= temp_arr[:,0].min()) & (qe_l <= temp_arr[:,0].max())]
qe_sub = qe.loc[(qe_l >= 392) & (qe_l <= 900)]

re_sub = filt_trans.loc[(filt_l >= 392) & (filt_l <= 900)]
qe_sub.shape, re_sub.shape
```

```python
temp = qe_sub.copy()
temp['QE_percent'] /= 100
ax = temp.plot(x='Wavelength_nm', y='QE_percent')
re_sub.plot(kind='line', x=re_sub.columns[0], y=re_sub.columns[1:],
             figsize=(15, 10), ax=ax)
```

try dividing out the spectral response curve and max normalizing

```python
# re_sub, temp
qe_vals = temp.values[:,1]
#qe_vals.shape #(509,)

rsr = re_sub.copy()
rsr[rsr.columns[1:]] = rsr[rsr.columns[1:]] * qe_vals[..., np.newaxis]
```

```python
rsr.plot(kind='line', x=rsr.columns[0], y=rsr.columns[1:],
             figsize=(15, 10))
```

```python
# peak normalize
rsr_pn = rsr.copy()
rel_max = rsr_pn[rsr_pn.columns[1:]].values.max(axis=0)
rel_max = rel_max[..., np.newaxis]
rsr_pn[rsr_pn.columns[1:]] = rsr_pn[rsr_pn.columns[1:]] / rel_max.T 
rsr_pn.plot(kind='line', x=rsr.columns[0], y=rsr.columns[1:],
             figsize=(15, 10))
```

```python
rename_dict = {}
new_cols = ['b', 'g', 'r', 're', 'nir']
for i, c in enumerate(new_cols):
    rename_dict[rsr_pn.columns.values[i+1]] = new_cols[i]

rename_dict
# rsr_pn.rename
```

```python
rsr_pn.rename(columns=rename_dict).to_csv('rededge_rsr.csv')
```
