```python
# generate some gaussians
from scipy.stats import norm

def gen_fake_gauss(mu, sigma, wavs):
    mu = 435 
    sigma = 1.5
    x1 = wavs.min()
    x2 = wavs.max()

    #Next, we calculate the Z-transform of the lower and upper bound using the mean and standard deviation defined above.

    # calculate the z-transform
    z1 = ( x1 - mu ) / sigma
    z2 = ( x2 - mu ) / sigma

    #After the Z-transform of the lower and upper bounds are calculated, we calculate the probability with SciPy's scipy.stats.norm.pdf() function.

    x = np.arange(z1, z2, 0.001) # range of x in spec
    x = np.arange(z1, z2) # range of x in spec
    x_all = np.arange(-10, 10, 0.001) # entire range of x, both in and out of spec
    x_all = l8_rsr_ca[cols[0]]
    # mean = 0, stddev = 1, since Z-transform was calculated
    y = norm.pdf(x,0,1)
    y2 = norm.pdf(x_all,0,1)

    return y

plt.plot(l8_rsr_ca[cols[0]], l8_rsr_ca[cols[1]]);
for m in [435, 445, 455]:
    
    samp = gen_fake_gauss(m, 1.5, l8_rsr_ca[cols[0]])
    plt.plot(l8_rsr_ca[cols[0]], samp);

plt.show()
```

```python
y1 = b.copy()
y2 = b_shift.copy()
x = l.copy()
z = y1-y2
dx = x[1:] - x[:-1]
cross_test = np.sign(z[:-1] * z[1:])

x_intersect = x[:-1] - dx / (z[1:] - z[:-1]) * z[:-1]
dx_intersect = - dx / (z[1:] - z[:-1]) * z[:-1]

areas_pos = abs(z[:-1] + z[1:]) * 0.5 * dx # signs of both z are same

areas_neg = 0.5 * dx_intersect * abs(z[:-1]) + 0.5 * (dx - dx_intersect) * abs(z[1:])

areas = np.where(cross_test < 0, areas_neg, areas_pos)
total_area = np.sum(areas)

total_area, area_b, np.trapz(areas_pos), np.trapz(areas_neg)
```
